#### 7. Dado el siguiente texto con tweets, utiliza expresiones regulares para extraer: hashtags (`#`), menciones (`@`) y URLs. Muestra los resultados para cada tweet.

```
tweets = [
    "¡El #MachineLearning está cambiando el mundo! @OpenAI lo sabe bien.",
```

````python
import re

tweets = [
    "¡El #MachineLearning está cambiando el mundo! @OpenAI lo sabe bien.",
    "Nuevo paper de #NLP disponible: https://arxiv.org/abs/1810.04805 vía @GoogleAI",
    "Aprendiendo #Python y #spaCy con @huggingface hoy 🚀",
]

for tweet in tweets:
    hashtags  = re.findall(r'#\w+', tweet)
    menciones = re.findall(r'@\w+', tweet)
    urls      = re.findall(r'https?://\S+', tweet)
    print(f"Tweet    : {tweet}")
    print(f"Hashtags : {hashtags}")
    print(f"Menciones: {menciones}")
    print(f"URLs     : {urls}")
    print()
````

#### 8. Aplica un pipeline de limpieza de texto sobre la siguiente reseña usando `re.sub`. Los pasos deben ser en orden:
1. Reemplazar URLs por el token `[URL]`
2. Reemplazar correos electrónicos por el token `[EMAIL]`
3. Reemplazar números de teléfono (formato `(XX) XXXX-XXXX`) por `[TELÉFONO]`
4. Eliminar caracteres no ASCII (emojis, símbolos especiales)
5. Normalizar espacios múltiples a uno solo
6. Convertir a minúsculas

```
reseña = """¡¡INCREÍBLE producto!! 😍🔥 Cómpralo en https://tienda.com/promo?ref=123
Contacto: soporte@tienda.com o llama al (55) 8765-4321
   Mucho   espacio   extra   aquí   """
```

````python
import re

reseña = """¡¡INCREÍBLE producto!! 😍🔥 Cómpralo en https://tienda.com/promo?ref=123
Contacto: soporte@tienda.com o llama al (55) 8765-4321
   Mucho   espacio   extra   aquí   """

texto_limpio = reseña

# Paso 1: Reemplazar URLs
texto_limpio = re.sub(r'https?://\S+', '[URL]', texto_limpio)

# Paso 2: Reemplazar correos
texto_limpio = re.sub(r'[\w.]+@[\w.]+\.[a-zA-Z]{2,}', '[EMAIL]', texto_limpio)

# Paso 3: Reemplazar teléfonos formato (XX) XXXX-XXXX
texto_limpio = re.sub(r'\(\d{2}\)\s\d{4}-\d{4}', '[TELÉFONO]', texto_limpio)

# Paso 4: Eliminar caracteres no ASCII (emojis, símbolos especiales)
texto_limpio = re.sub(r'[^\x00-\x7Fáéíóúüñ¡¿ÁÉÍÓÚ \n.,!?@()+-]', '', texto_limpio)

# Paso 5: Normalizar espacios múltiples
texto_limpio = re.sub(r'\s+', ' ', texto_limpio).strip()

# Paso 6: Convertir a minúsculas
texto_limpio = texto_limpio.lower()

print("Original:")
print(reseña)
print("\nLimpio:")
print(texto_limpio)
````

## Módulo `string`

#### 9. Usa el módulo `string` para limpiar la puntuación del siguiente texto. Realiza los siguientes pasos:
1. Elimina toda la puntuación (incluye `¡` y `¿`) usando `str.maketrans` y `.translate()`
2. Convierte el resultado a minúsculas
3. Divide el texto en una lista de palabras (tokens) usando `.split()`
4. Imprime: el texto limpio, la lista de tokens y el número total de tokens

```
texto = "¡Hola, mundo! ¿Cómo estás? El NLP... es fascinante."
```

````python
import string

texto = "¡Hola, mundo! ¿Cómo estás? El NLP... es fascinante."

# Paso 1: Eliminar puntuación con str.maketrans y .translate()
tabla = str.maketrans('', '', string.punctuation + '¡¿')
texto_sin_puntuacion = texto.translate(tabla)

# Paso 2: Convertir a minúsculas
texto_lower = texto_sin_puntuacion.lower()

# Paso 3: Dividir en tokens
tokens = texto_lower.split()

# Paso 4: Imprimir resultados
print("Texto limpio  :", texto_lower)
print("Tokens        :", tokens)
print("Total tokens  :", len(tokens))
````

````python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

parrafo = """El Procesamiento de Lenguaje Natural es una rama de la Inteligencia Artificial.
Permite que las computadoras lean y comprendan el lenguaje humano.
Sus aplicaciones incluyen chatbots, traducción automática y análisis de sentimientos."""

# Tokenización por oraciones
oraciones = sent_tokenize(parrafo, language='spanish')
print(f"Número de oraciones: {len(oraciones)}")
for i, oracion in enumerate(oraciones, 1):
    print(f"  Oración {i}: {oracion.strip()}")

# Tokenización por palabras
tokens = word_tokenize(parrafo, language='spanish')
print(f"\nNúmero de word tokens: {len(tokens)}")
print(f"Tokens: {tokens}")
````

## Conceptos Fundamentales de NLP

#### 10. Dado el siguiente mini-corpus, responde las preguntas **con código**:

```
corpus = [
    "el gato duerme en la alfombra",
    "el perro corre en el parque",
    "la gata y el perro son amigos",
    "el parque tiene una alfombra verde",
]
```

a) Calcula el **vocabulario** del corpus (tokens únicos en minúsculas, sin puntuación). ¿Cuántos tipos (`types`) hay?

b) ¿Cuántos tokens en total hay en todo el corpus?

c) Calcula la **frecuencia** de cada tipo. ¿Cuáles son las 5 palabras más frecuentes? ¿Se observa la **Ley de Zipf**?

d) ¿Qué palabras del vocabulario podrían considerarse **stopwords**? Identifícalas filtrando con NLTK y muestra el vocabulario resultante sin ellas.

````python
from collections import Counter
from nltk.corpus import stopwords

corpus = [
    "el gato duerme en la alfombra",
    "el perro corre en el parque",
    "la gata y el perro son amigos",
    "el parque tiene una alfombra verde",
]

# a) Vocabulario: tokens únicos
todos_los_tokens = [token for oracion in corpus for token in oracion.split()]
vocabulario = set(todos_los_tokens)
print(f"a) Vocabulario ({len(vocabulario)} tipos): {sorted(vocabulario)}")

# b) Total de tokens
total_tokens = len(todos_los_tokens)
print(f"\nb) Total de tokens en el corpus: {total_tokens}")

# c) Frecuencias — Ley de Zipf
frecuencias = Counter(todos_los_tokens)
print(f"\nc) Top 5 palabras más frecuentes:")
for palabra, freq in frecuencias.most_common(5):
    print(f"   '{palabra}': {freq}")

# d) Vocabulario sin stopwords
STOPWORDS_ES = set(stopwords.words('spanish'))
vocabulario_limpio = {w for w in vocabulario if w not in STOPWORDS_ES}
print(f"\nd) Vocabulario sin stopwords ({len(vocabulario_limpio)} tipos): {sorted(vocabulario_limpio)}")
````
